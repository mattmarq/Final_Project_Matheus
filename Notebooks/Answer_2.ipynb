```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("data.csv")
```

```python
df
```

```python
df["R_total"] = df.R_wins + df.R_losses + df.R_draw
```

```python
df["B_total"] = df.B_wins + df.B_losses + df.B_draw
```

```python
df["R_eff"] = df.R_wins / df.R_total
```

```python
df["B_eff"] = df.B_wins / df.B_total
```

```python
df = df.sort_values(by=["B_eff","B_total"], ascending=False)
```
